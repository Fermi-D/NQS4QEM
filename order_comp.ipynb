```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# space complexity
fig, ax = plt.subplots(figsize=(7,5))
qubit_range = np.linspace(1, 100, 100)

c1,c2,c3,c4 = "red","blue","green","orange"
l1,l2,l3,l4 = "density matrix","max eigen vector","state vector","virtual distillation (m=10)"

ax.set_xlabel('Number of Qubit')  # x axis title
ax.set_ylabel('Space Complexity')  # y axis title
#plt.yscale('log', basey=10)

ax.plot(qubit_range, 2**(2*qubit_range), linestyle = "dotted", linewidth = 4, color=c1, label=l1)
ax.plot(qubit_range, 2**(2*qubit_range), linestyle = "dotted", linewidth = 4, color=c2, label=l2)
ax.plot(qubit_range, 2**(qubit_range+1), linestyle = "dotted", linewidth = 4, color=c3, label=l3)
ax.plot(qubit_range, 10*qubit_range, linestyle = "dotted", linewidth = 4, color=c4, label=l4)

plt.grid(which='major', color='black', linestyle='-', alpha=0.2)
plt.grid(which='minor', color='black', linestyle='-', alpha=0.2)
plt.xlabel(r"Number of Qubit", fontsize=16)
plt.ylabel(r"Space Complexity", fontsize=16)
plt.yscale("log")
ax.legend(fontsize=16, loc=0)

fig.tight_layout()
fig.savefig("./space_complexity.pdf")
```

```python
# time complexity
fig, ax = plt.subplots(figsize=(7,5))
qubit_range = np.linspace(1, 100, 100)

c1,c2,c3,c4 = "red","blue","green","orange"
l1,l2,l3,l4 = "density matrix","max eigen vector","state vector","virtual distillation (m=10)"

ax.set_xlabel('Number of Qubit')  # x axis title
ax.set_ylabel('Time Complexity')  # y axis title
#plt.yscale('log', basey=10)

ax.plot(qubit_range, 2**(3*qubit_range), linestyle = "dotted", linewidth = 4, color=c1, label=l1)
ax.plot(qubit_range, 2**(3*qubit_range), linestyle = "dotted", linewidth = 4, color=c2, label=l2)
ax.plot(qubit_range, 2**(2*qubit_range), linestyle = "dotted", linewidth = 4, color=c3, label=l3)
ax.plot(qubit_range, 10*qubit_range, linestyle = "dotted", linewidth = 4, color=c4, label=l4)

plt.grid(which='major', color='black', linestyle='-', alpha=0.2)
plt.grid(which='minor', color='black', linestyle='-', alpha=0.2)
plt.xlabel(r"Number of Qubit", fontsize=16)
plt.ylabel(r"Time Complexity", fontsize=16)
plt.yscale("log")
ax.legend(fontsize=16, loc=0)

fig.tight_layout()
fig.savefig("./time_complexity.pdf")
```
