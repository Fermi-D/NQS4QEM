```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_copy_list = np.arange(1, 11)
n_copy_list = np.sort(n_copy_list)[::-1]
n_sampling_list = [10, 50, 100, 1000]
abs_error_list = []
abs_error_array = []
var_list = []
var_array = []

for n_copy in n_copy_list:
    for n_sampling in n_sampling_list:
        df = pd.read_csv(f"./train_log_n_copy={n_copy}_n_sample={n_sampling}.csv")
        abs_error_list.append(np.abs(1-df["Observable_XX_ev"].values[-1]))
        var_list.append(df["Observable_XX_var"].values[-1])
    abs_error_array.append(abs_error_list)
    var_array.append(var_list)
```

```python
# absolute error  
fig, ax = plt.subplots(figsize=(10, 10))
im = ax.imshow(abs_error_array)

ax.set_xticks(np.arange(len(n_sampling_list)), labels=np.array(n_sampling_list).astype("str"))
ax.set_yticks(np.arange(len(n_copy_list)), labels=np.array(n_copy_list).astype("str"))

ax.set_title("absolute error (depolarizing & unitary error)")
fig.tight_layout()
fig.colorbar(im, ax=ax)
plt.show()
```

```python
# variance
# absolute error  
fig, ax = plt.subplots(figsize=(10, 10))
im = ax.imshow(var_array)

ax.set_xticks(np.arange(len(n_sampling_list)), labels=np.array(n_sampling_list).astype("str"))
ax.set_yticks(np.arange(len(n_copy_list)), labels=np.array(n_copy_list).astype("str"))

ax.set_title("variance (depolarizing & unitary error)")
fig.tight_layout()
fig.colorbar(im, ax=ax)
plt.show()
```
